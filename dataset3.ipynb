```python
import pandas as pd
import plotly.express as px
#dataset link https://www.kaggle.com/sobhanmoosavi/us-accidents
```

```python
df = pd.read_csv('datasets/crash.csv') #reading in our csv as a dataframe
df
```

```python
df1 = df[(df.State == 'TN')| (df.State == 'NC')|(df.State == 'GA')] #we only want crashes in NC, TN, and GA
df1 = df1[(df1.Severity == 4)] #we only want the crashes in TN,GA,NC to have severity of 4
df1
```

```python
#plotting using mapbox and plotly
fig = px.scatter_mapbox(df1, lat='Start_Lat', lon='Start_Lng', hover_name="Description", title = 'Car Crashes that Occured in TN, NC, GA in 2016 - 2020 with Severity of 4', 
                        center=dict(lat=35, lon=-85), zoom=5.5,
                       mapbox_style="open-street-map")
                       
fig.show()
```

```python
groups = df1.groupby(['Severity','State']).size() #i also wanted to see the count of the amount of severity 4 crashes, i made a 'group' of severity and state
groups.plot.bar()
```
